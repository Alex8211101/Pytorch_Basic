```python
import torch
import torch.nn.functional as F
```

```python
x_data = torch.Tensor([[1.0], [2.0], [3.0]])
y_data = torch.Tensor([[0.0], [0.0], [1.0]])
```

#### ----------------------------------------------------------------------------------------------------------------------

```python
class LogisticRegressionModel(torch.nn.Module):
    def __init__(self):
        super(LogisticRegressionModel, self).__init__()
        self.linear = torch.nn.Linear(1, 1)
        
    def forward(self, x):
        y_pred = F.sigmoid(self.linear(x))
        return y_pred
```

```python
model = LogisticRegressionModel()
```

#### ----------------------------------------------------------------------------------------------------------------------

```python
%%html
<img src='picture/loss.jpg', width=400>
```

```python
criterion = torch.nn.BCELoss(size_average=False)      # loss 函数表达式如上
optimizer = torch.optim.SGD(model.parameters(), lr = 0.1)
```

#### ----------------------------------------------------------------------------------------------------------------------

```python
for epoch in range(1000):
    y_pred = model(x_data)
    loss = criterion(y_pred, y_data)
    print(epoch, loss.item())
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
print("w = ", model.linear.weight.item())
print("b = ", model.linear.bias.item())
```

#### ----------------------------------------------------Result-------------------------------------------------------------

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 200)           # x 在 [0, 10) 上平均点样 200 个点
x_t = torch.Tensor(x).view((200, 1))  # 将 x 转换成 200 × 1 的矩阵, 相当于numpy中的reshape()
y_t = model(x_t)
y = y_t.data.numpy()                  # 将训练得到的tensor格式的数组转换为numpy格式用于绘图
plt.plot(x, y)
plt.plot([0, 10],[0.5, 0.5], c='r')   # 做一条红线
plt.xlim([0, 10])
plt.xlabel('Hours')
plt.ylabel('Probability of Pass')
plt.grid()                            # 绘制网格
plt.show()
```
